```python
import math

def uav_wing_sizing(
    mass=17,                 # kg
    V_stall=14,              # m/s
    CL_max=1.4,              # stall CL (from airfoil, NOT guessed randomly)
    CL_cruise=0.6,           # cruise CL (design choice)
    AR=8.2,
    taper=0.45,
    rho=1.225,
    mu=1.81e-5
):
    g = 9.81

    # ---------------------------
    # 1. Weight
    # ---------------------------
    W = mass * g

    # ---------------------------
    # 2. Wing loading (STALL ONLY)
    # ---------------------------
    W_S = 0.5 * rho * V_stall**2 * CL_max

    # ---------------------------
    # 3. Wing area
    # ---------------------------
    S = W / W_S

    # ---------------------------
    # 4. Geometry
    # ---------------------------
    b = math.sqrt(AR * S)

    cr = (2 * S) / (b * (1 + taper))
    ct = taper * cr

    MAC = (2/3) * cr * ((1 + taper + taper**2) / (1 + taper))
    y_mac = (b / 6) * ((1 + 2*taper) / (1 + taper))

    # ---------------------------
    # 5. Cruise condition
    # ---------------------------
    V_cruise = math.sqrt((2 * W) / (rho * S * CL_cruise))

    # ---------------------------
    # 6. Reynolds number
    # ---------------------------
    Re = (rho * V_cruise * MAC) / mu

    # ---------------------------
    # DEBUG + OUTPUT
    # ---------------------------
    print("\n===== UAV Wing Design =====")

    print("\n--- INPUT CHECK ---")
    print(f"CL_max (STALL): {CL_max}")
    print(f"CL_cruise: {CL_cruise}")
    print(f"V_stall: {V_stall} m/s")

    print("\n--- WEIGHT ---")
    print(f"Mass: {mass} kg")
    print(f"Weight: {W:.2f} N")

    print("\n--- WING LOADING (from stall) ---")
    print(f"W/S: {W_S:.2f} N/m^2  <-- THIS must be ~235")

    print("\n--- GEOMETRY ---")
    print(f"Wing Area (S): {S:.3f} m^2")
    print(f"Wingspan (b): {b:.3f} m")
    print(f"Root Chord (cr): {cr:.3f} m")
    print(f"Tip Chord (ct): {ct:.3f} m")
    print(f"MAC: {MAC:.3f} m")
    print(f"MAC Position: {y_mac:.3f} m")

    print("\n--- PERFORMANCE ---")
    print(f"Cruise Speed: {V_cruise:.2f} m/s")
    print(f"Reynolds Number: {Re:.2e}")

    return {
        "W/S": W_S,
        "S": S,
        "b": b,
        "cr": cr,
        "ct": ct,
        "MAC": MAC,
        "Re": Re,
        "V_cruise": V_cruise
    }


# RUN
uav_wing_sizing()
```
