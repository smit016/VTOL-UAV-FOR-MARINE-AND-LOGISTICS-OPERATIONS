```python
import math

def first_layer_height(U, L, rho=1.225, mu=1.789e-5, y_plus=1.0):
    """
    Calculate first layer height for a given y+.

    Parameters:
    U       : Free stream velocity (m/s)
    L       : Characteristic length (m) -> use MAC or smaller chord
    rho     : Density (kg/m^3)
    mu      : Dynamic viscosity (kg/m·s)
    y_plus  : Target y+ value (default = 1)

    Returns:
    dict with Re, Cf, tau_w, u_tau, y (first layer height)
    """

    # 1. Reynolds number
    Re = (rho * U * L) / mu

    # 2. Skin friction coefficient (turbulent flat plate approximation)
    Cf = 0.026 * (Re ** (-1/7))

    # 3. Wall shear stress
    tau_w = 0.5 * rho * U**2 * Cf

    # 4. Friction velocity
    u_tau = math.sqrt(tau_w / rho)

    # 5. First layer height
    y = (y_plus * mu) / (rho * u_tau)

    return {
        "Re": Re,
        "Cf": Cf,
        "tau_w": tau_w,
        "u_tau": u_tau,
        "first_layer_height_m": y,
        "first_layer_height_mm": y * 1000
    }


# ---- YOUR CASE ----
U = 25        # m/s
L = 0.3       # m (MAC)
y_plus_target = 1

result = first_layer_height(U, L, y_plus=y_plus_target)

print("----- CFD Wall Resolution Calculation -----")
print(f"Reynolds Number       : {result['Re']:.2e}")
print(f"Skin Friction Coeff   : {result['Cf']:.5f}")
print(f"Wall Shear Stress (Pa): {result['tau_w']:.4f}")
print(f"Friction Velocity     : {result['u_tau']:.4f} m/s")
print(f"First Layer Height    : {result['first_layer_height_m']:.2e} m")
print(f"                     = {result['first_layer_height_mm']:.4f} mm")
```
