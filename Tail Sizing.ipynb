```python
import math

class UAVTailDesign:
    def __init__(self,
                 Sw=0.71,        # wing area (m^2)
                 b=2.41,         # wingspan (m)
                 c=0.295,        # mean aerodynamic chord (m)
                 lt=0.95,        # tail arm (m)
                 Vh=0.6,         # horizontal tail volume coeff
                 Vv=0.05,        # vertical tail volume coeff
                 ARh=4.0,        # horizontal tail aspect ratio
                 ARv=1.5,        # vertical tail aspect ratio
                 twin_tail=True  # twin boom config
                 ):

        self.Sw = Sw
        self.b = b
        self.c = c
        self.lt = lt
        self.Vh = Vh
        self.Vv = Vv
        self.ARh = ARh
        self.ARv = ARv
        self.twin_tail = twin_tail

    # -------------------------
    # Horizontal Tail
    # -------------------------
    def horizontal_tail_area(self):
        Sh = (self.Vh * self.Sw * self.c) / self.lt
        return Sh

    def horizontal_tail_geometry(self):
        Sh = self.horizontal_tail_area()
        bh = math.sqrt(self.ARh * Sh)
        ch = Sh / bh
        return Sh, bh, ch

    # -------------------------
    # Vertical Tail
    # -------------------------
    def vertical_tail_area(self):
        Sv = (self.Vv * self.Sw * self.b) / self.lt
        return Sv

    def vertical_tail_geometry(self):
        Sv = self.vertical_tail_area()

        if self.twin_tail:
            Sv_each = Sv / 2
        else:
            Sv_each = Sv

        bv = math.sqrt(self.ARv * Sv_each)
        cv = Sv_each / bv

        return Sv, Sv_each, bv, cv

    # -------------------------
    # Control Surface Sizing
    # -------------------------
    def control_surfaces(self):
        _, _, ch = self.horizontal_tail_geometry()
        _, _, _, cv = self.vertical_tail_geometry()

        elevator = (0.25 * ch, 0.30 * ch)
        rudder = (0.30 * cv, 0.40 * cv)

        return elevator, rudder

    # -------------------------
    # Summary Print
    # -------------------------
    def summary(self):
        print("\n--- UAV Tail Design Summary ---")

        Sh, bh, ch = self.horizontal_tail_geometry()
        print(f"\nHorizontal Tail:")
        print(f"Area Sh = {Sh:.3f} m^2")
        print(f"Span bh = {bh:.3f} m")
        print(f"Chord ch = {ch:.3f} m")

        Sv, Sv_each, bv, cv = self.vertical_tail_geometry()
        print(f"\nVertical Tail:")
        print(f"Total Area Sv = {Sv:.3f} m^2")
        print(f"Each Tail Area = {Sv_each:.3f} m^2")
        print(f"Height bv = {bv:.3f} m")
        print(f"Chord cv = {cv:.3f} m")

        elevator, rudder = self.control_surfaces()
        print(f"\nControl Surfaces:")
        print(f"Elevator chord = {elevator[0]:.3f} to {elevator[1]:.3f} m")
        print(f"Rudder chord = {rudder[0]:.3f} to {rudder[1]:.3f} m")


# -------------------------
# Run with your values
# -------------------------
if __name__ == "__main__":
    uav = UAVTailDesign(
        Sw=0.71,
        b=2.41,
        c=0.295,
        lt=0.95,
        Vh=0.6,
        Vv=0.05,
        ARh=4.0,
        ARv=1.5,
        twin_tail=True
    )

    uav.summary()
```
